```python
def fill_age(row):
    grouped_train = train.groupby(['temper_rolling','strength'])
    grouped_median_train = grouped_train.median()
    grouped_median_train = grouped_median_train.reset_index()[['temper_rolling','strength','surface-quality']]
    print("nadia")
    print(grouped_median_train.shape)
    grouped_median_train.head(131)
    condition = (
        (grouped_median_train['temper_rolling'] == row['temper_rolling']) &
        (grouped_median_train['strength'] == row['strength']) 
    )
    print(grouped_median_train[condition]['surface-quality'].values)
    if not np.isnan(grouped_median_train[condition]['surface-quality'].values):
        return grouped_median_train[condition]['surface-quality'].mode().iloc[0]
    elif row['strength'] > 0:
        return 4
```

```python
train['surface-quality'] = train.apply(lambda row: fill_age(row) if np.isnan(row['surface-quality']) else row['surface-quality'], axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.grid_search import GridSearchCV
from sklearn.grid_search import RandomizedSearchCV
from sklearn.cross_validation import  cross_val_score
from time import time
from operator import itemgetter
from scipy.stats import randint
def report(grid_scores, n_top=3):
    """Report top n_top parameters settings, default n_top=3.

    Args
    ----
    grid_scores -- output from grid or random search
    n_top -- how many to report, of top models

    Returns
    -------
    top_params -- [dict] top parameter settings found in
                  search
    """
    top_scores = sorted(grid_scores,
                        key=itemgetter(1),
                        reverse=True)[:n_top]
    for i, score in enumerate(top_scores):
        print("Model with rank: {0}".format(i + 1))
        print(("Mean validation score: "
               "{0:.3f} (std: {1:.3f})").format(
               score.mean_validation_score,
               np.std(score.cv_validation_scores)))
        print("Parameters: {0}".format(score.parameters))
        print("")

    return top_scores[0].parameters

def run_gridsearch(X, y, clf, param_grid, cv=5):
    """Run a grid search for best Decision Tree parameters.

    Args
    ----
    X -- features
    y -- targets (classes)
    cf -- scikit-learn Decision Tree
    param_grid -- [dict] parameter settings to test
    cv -- fold of cross-validation, default 5

    Returns
    -------
    top_params -- [dict] from report()
    """
    grid_search = GridSearchCV(clf,
                               param_grid=param_grid,
                               cv=cv)
    start = time()
    grid_search.fit(X, y)

    print(("\nGridSearchCV took {:.2f} "
           "seconds for {:d} candidate "
           "parameter settings.").format(time() - start,
                len(grid_search.grid_scores_)))

    top_params = report(grid_search.grid_scores_, 3)
    return  top_params
param_grid = {"criterion": ["gini", "entropy"],
              "min_samples_split": [2, 10, 20],
              "max_depth": [None, 2, 5, 10],
              "min_samples_leaf": [1, 5, 10],
              "max_leaf_nodes": [None, 5, 10, 20],
              }

dt = DecisionTreeClassifier()
train = train.reset_index(drop=True)

y_train = train['classes']
print(y_train.head())

tt=train.drop(['classes'],axis=1)
tt=tt.drop(['class_1'], axis=1)
tt=tt.drop(['class_2'], axis=1)
tt=tt.drop(['class_3'], axis=1)
tt=tt.drop(['class_5'], axis=1)
tt=tt.drop(['class_U'], axis=1)
tt = tt.reset_index(drop=True)
x_train = tt.values
print(x_train)

ts_gs = run_gridsearch(x_train, y_train, dt, param_grid, cv=10)
```
