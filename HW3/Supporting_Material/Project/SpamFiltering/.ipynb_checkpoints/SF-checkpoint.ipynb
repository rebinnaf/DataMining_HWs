```python
import warnings
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore', category=DeprecationWarning)

import pandas as pd
from sklearn.cross_validation import train_test_split

pd.options.display.max_columns = 100

from matplotlib import pyplot as plt
import numpy as np

import seaborn as sns

import pylab as plot
dataset = pd.read_csv('spam.csv',encoding='latin-1')
dataset.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
dataset = dataset.rename(columns={'v1': 'class','v2': 'text'})
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem import SnowballStemmer
from nltk.corpus import stopwords
import string
def do_stemming(text):
    stemmed = ''
    nopunc = [char for char in text if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    for word in nopunc.split():
        stemmed = stemmed+' '+PorterStemmer().stem(word)


        # stemmed.append(LancasterStemmer().stem(f))
        # stemmed.append(SnowballStemmer('english').stem(f))
    return stemmed
def process_text(text):
    '''
    What will be covered:
    1. Remove punctuation
    2. Remove stopwords
    3. Return list of clean text words
    '''
    
    #1
    nopunc = [char for char in text if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    
    #2
    clean_words = [SnowballStemmer('english').stem(word) for word in nopunc.split() if word.lower() not in stopwords.words('english')]
    #3

    return clean_words
```

```python
dataset['text']=dataset['text'].apply(process_text)
dataset.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_transformer = CountVectorizer(analyzer=process_text)
bow_transformer.fit(dataset['text'])
bow_transformer.get_feature_names()
```
